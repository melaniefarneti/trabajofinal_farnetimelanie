### Nulos

```python
# Agrupar por 'SEQN'
dieta = dieta.groupby('id_paciente').agg(
    lambda x: x.sum() if x.dtype in [np.float64, np.int64] else x.first()
).reset_index()

# Para la columna 'Año', mantener el primer valor
dieta['año'] = dieta.groupby('id_paciente')['año'].first().values
dieta
```

```python
dieta.isnull().sum()
```

```python
# Identificar columnas a verificar (excluyendo 'id_paciente' y 'año')
columns_to_check = dieta.columns.difference(['id_paciente', 'año'])

# Filtrar filas donde todas las columnas relevantes son nulas
rows_with_all_nulls = dieta[columns_to_check].isnull().all(axis=1)

# Verificar cuántas filas tienen todas las columnas nulas
print(f"Filas a eliminar: {rows_with_all_nulls.sum()}")

# Eliminar las filas completamente nulas en las columnas relevantes
dieta = dieta[~rows_with_all_nulls]
dieta
```
