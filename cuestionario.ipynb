### Ingeniería de características

#### Actividad Física

```python
# Definir las columnas de actividad vigorosa y moderada
columnas_vigorosa = ['PAD200', 'PAQ605', 'PAQ650']  # Actividad vigorosa
columnas_moderada = ['PAD320', 'PAQ620', 'PAQ635', 'PAQ665', 'PAD020']  # Actividad moderada, incluyendo bicicleta

def determinar_actividad(row):
    # Determinar la presencia de actividad en trabajo y tiempo libre
    actividad_vigorosa_trabajo = row.get('PAQ605', 2) == 1  # Trabajo
    actividad_vigorosa_tiempo_libre_1 = row.get('PAD200', 2) == 1  # Tiempo libre
    actividad_vigorosa_tiempo_libre_2 = row.get('PAQ650', 2) == 1  # Tiempo libre

    actividad_moderada_trabajo = row.get('PAQ620', 2) == 1  # Trabajo
    actividad_moderada_tiempo_libre_1 = row.get('PAD320', 2) == 1  # Tiempo libre
    actividad_moderada_tiempo_libre_2 = row.get('PAQ665', 2) == 1  # Tiempo libre
    bicicleta_1 = row.get('PAD020', 2) == 1
    bicicleta_2 = row.get('PAQ635', 2) == 1  # Bicicleta para transportarse

    # Lógica para determinar el nivel de actividad
    if (
        actividad_vigorosa_trabajo
        or actividad_vigorosa_tiempo_libre_1
        or actividad_vigorosa_tiempo_libre_2
    ):
        return 1  # Actividad vigorosa
    elif (
        actividad_moderada_trabajo
        or actividad_moderada_tiempo_libre_1
        or actividad_moderada_tiempo_libre_2
        or bicicleta_1
        or bicicleta_2
    ):
        return 2  # Actividad moderada
    else:
        # Verificar si no hay actividad reportada en las columnas de interés
        if all(row.get(col, 2) == 2 for col in columnas_vigorosa + columnas_moderada):
            return 3  # No hay actividad
        else:
            # Si los datos son insuficientes o rechazados, devolver nulo
            if any(row.get(col) is None or row.get(col, 2) == 9 for col in columnas_vigorosa + columnas_moderada):
                return None  # Datos insuficientes o rechazados

# Recorrer cada DataFrame en el diccionario
for year, df in dataframes.items():
    # Aplicar la función determinar_actividad y agregar la nueva columna 'Actividad'
    df['Actividad'] = df.apply(determinar_actividad, axis=1)

    # Eliminar las columnas originales de actividad si están presentes en el DataFrame
    columnas_a_eliminar = [col for col in columnas_vigorosa + columnas_moderada if col in df.columns]
    df.drop(columns=columnas_a_eliminar, inplace=True, errors='ignore')
```

#### Diabetes

```python
# Función para procesar la columna DIQ010
def procesar_diabetes(row):
    valor = row.get('DIQ010', np.nan)
    if valor in [1, 2, 3]:  # Valores válidos
        return valor
    elif valor in [7, 9]:  # Rechazado o No lo sé -> NaN
        return np.nan
    else:  # Valores faltantes permanecen como NaN
        return valor

# Recorrer cada DataFrame en el diccionario
for year, df in dataframes.items():
    if 'DIQ010' in df.columns:
        # Procesar la columna y crear la nueva columna 'Diabetes'
        df['Diabetes'] = df.apply(procesar_diabetes, axis=1)

        # Eliminar la columna original 'DIQ010'
        df.drop(columns=['DIQ010'], inplace=True, errors='ignore')
```

#### Salud bucal

```python
# Función para inferir salud bucal en años sin variables directas
def inferir_salud_bucal(row):
    # Evaluar necesidades específicas
    limpieza_necesaria = row.get('OHD750', 0) == 1
    alivio_dolor = row.get('OHD740', 0) == 1
    tratamiento_encias = row.get('OHD730', 0) == 1
    extracciones_necesarias = any(row.get(var, 0) in [1, 2] for var in [
        'OHD711A', 'OHD711B', 'OHD711C', 'OHD711D', 'OHD711E', 'OHD711G'
    ])

    # Asignar códigos según las reglas definidas
    if not limpieza_necesaria and not alivio_dolor and not tratamiento_encias and not extracciones_necesarias:
        return 1  # Excelente
    elif limpieza_necesaria and not (alivio_dolor or tratamiento_encias or extracciones_necesarias):
        return 2  # Muy bien
    elif alivio_dolor or tratamiento_encias:
        return 3  # Bien
    elif extracciones_necesarias and not (alivio_dolor or tratamiento_encias):
        return 4  # Justo
    elif extracciones_necesarias and (alivio_dolor or tratamiento_encias):
        return 5  # Pobre
    else:
        return np.nan  # Casos faltantes o no válidos

# Procesar cada DataFrame
for year, df in dataframes.items():
    if 'OHQ010' in df.columns:  # Año 2000
        # Renombrar OHQ010 a 'Salud bucal' y reemplazar valores no válidos
        df.rename(columns={'OHQ010': 'Salud bucal'}, inplace=True)
        df['Salud bucal'].replace({7: np.nan, 9: np.nan}, inplace=True)

    elif 'OHQ011' in df.columns:  # Años con OHQ011
        # Renombrar OHQ011 a 'Salud bucal' y reemplazar valores no válidos
        df.rename(columns={'OHQ011': 'Salud bucal'}, inplace=True)
        df['Salud bucal'].replace({11: 1, 12: 2, 13: 3, 14: 4, 15: 5,  77: np.nan, 99: np.nan}, inplace=True)

    elif 'OHQ845' in df.columns:  # Años con OHQ845
        # Renombrar OHQ845 a 'Salud bucal' y reemplazar valores no válidos
        df.rename(columns={'OHQ845': 'Salud bucal'}, inplace=True)
        df['Salud bucal'].replace({7: np.nan, 9: np.nan}, inplace=True)

    elif any(col in df.columns for col in ['OHD711A', 'OHD750']):  # Años con inferencia
        # Crear 'Salud bucal' basado en inferencia
        df['Salud bucal'] = df.apply(inferir_salud_bucal, axis=1)

        # Eliminar columnas originales relacionadas con la inferencia
        columnas_a_eliminar = [
            'OHD711A', 'OHD711B', 'OHD711C', 'OHD711D', 'OHD711E', 'OHD711G',
            'OHD721', 'OHD730', 'OHD740', 'OHD750'
        ]
        df.drop(columns=[col for col in columnas_a_eliminar if col in df.columns], inplace=True, errors='ignore')
```

#### Tabaquismo

```python
demo = pd.read_csv('/content/demografico.csv')

def unificar_tabaquismo(row, demo):
    """
    Unifica las columnas SMQ620 y SMQ040 en una sola variable 'Tabaquismo'
    """
    edad = demo.loc[demo['id_paciente'] == row['SEQN'], 'edad'].values[0]  # Obtener la edad del participante según 'seqn'
    tabaquismo_12_19 = row.get('SMQ620', np.nan)  # Tabaquismo 12-19 años
    tabaquismo_mayor_20 = row.get('SMQ040', np.nan)  # Tabaquismo >20 años

    # Si la edad es menor a 12 años, asignar 'No' directamente
    if edad < 12:
        return '2'

    # Evaluar tabaquismo para 12-19 años
    if tabaquismo_12_19 == 1:  # 1 = Sí
        return '1'
    elif tabaquismo_12_19 == 2:  # 2 = No
        return '2'
    elif tabaquismo_12_19 in [7, 9]:  # 7 = Rechazado, 9 = No lo sé
        tabaquismo_12_19 = np.nan

    # Evaluar tabaquismo para >20 años
    if tabaquismo_mayor_20 in [1, 2]:  # 1 = Cada día, 2 = Algunos días
        return '1'
    elif tabaquismo_mayor_20 == 3:  # 3 = No
        return '2'
    elif tabaquismo_mayor_20 in [7, 9]:  # 7 = Rechazado, 9 = No lo sé
        tabaquismo_mayor_20 = np.nan

    # Unificación: determinar resultado final basado en ambas columnas
    if tabaquismo_12_19 == 'Sí' or tabaquismo_mayor_20 == 'Sí':
        return '1'
    elif tabaquismo_12_19 == 'No' and tabaquismo_mayor_20 == 'No':
        return '2'
    else:
        return np.nan  # Si ambos son inválidos o faltantes, asignar NaN

# Aplicar a todos los DataFrames
for year, df in dataframes.items():
    if 'SMQ620' in df.columns or 'SMQ040' in df.columns:
        # Crear la columna 'Tabaquismo' unificada
        df['Tabaquismo'] = df.apply(lambda row: unificar_tabaquismo(row, demo), axis=1)

        # Eliminar columnas originales relacionadas con tabaquismo
        columnas_a_eliminar = ['SMQ620', 'SMQ040']
        df.drop(columns=[col for col in columnas_a_eliminar if col in df.columns], inplace=True, errors='ignore')
```

#### Drogas

```python
# Función para unificar las variables DUQ200 y DUQ240 en 'Drogas'
def procesar_drogas(row, edad_col='edad'):
    # Obtener la edad del participante
    edad = demo.loc[demo['id_paciente'] == row['SEQN'], 'edad'].values[0]  # Obtener la edad del participante según 'seqn'

    # Si la edad es menor a 12 años, asignar "No" directamente
    if edad < 12:
        return 2  # 2 = No

    # Obtener los valores de DUQ200 (Marihuana) y DUQ240 (Cocaína)
    marihuana = row.get('DUQ200', np.nan)
    cocaina = row.get('DUQ240', np.nan)

    # Si cualquiera de las variables es 1 (Sí), devolver 1 (Sí)
    if marihuana == 1 or cocaina == 1:
        return 1  # Sí
    # Si ambas variables son 2 (No), devolver 2 (No)
    elif marihuana == 2 and cocaina == 2:
        return 2  # No
    # Si cualquiera es 7 (Rechazado) o 9 (No lo sé), devolver NaN
    elif marihuana in [7, 9] or cocaina in [7, 9]:
        return np.nan
    # Si ambas son NaN, devolver NaN
    else:
        return np.nan

# Recorrer cada DataFrame en el diccionario
for year, df in dataframes.items():
    if 'DUQ100' in df.columns:  # Año 2000, la variable ya está unificada
        # Renombrar DUQ100 a 'Drogas' directamente
        df.rename(columns={'DUQ100': 'Drogas'}, inplace=True)
    elif 'DUQ200' in df.columns or 'DUQ240' in df.columns:  # Años posteriores
        # Crear la columna 'Drogas' unificando DUQ200 y DUQ240
        df['Drogas'] = df.apply(procesar_drogas, axis=1)

        # Eliminar las columnas originales DUQ200 y DUQ240
        columnas_a_eliminar = [col for col in ['DUQ200', 'DUQ240'] if col in df.columns]
        df.drop(columns=columnas_a_eliminar, inplace=True, errors='ignore')
```

### Manejo de Nulos

```python
# Separación de features en numéricas, binarias y categóricas.
# Numéricas
num_feats = ['id_paciente', 'año']
# Categóricas binarias
bin_feats = ['Tabaquismo', 'Drogas']
# Categóricas Multi-
nom_feats = ['Salud bucal', 'Actividad', 'Diabetes']
cat_feats = nom_feats + bin_feats
```

```python
# Asegurarse de que las columnas contengan valores enteros
for column in cat_feats:
    # Reemplaza NaN por un valor temporal, asegurando que la conversión sea segura
    cuestionario[column] = cuestionario[column].fillna(-1)

    # Convertir a enteros
    cuestionario[column] = cuestionario[column].astype(int)
    cuestionario[column] = cuestionario[column].replace(-1, pd.NA)  # Cambia -1 de vuelta a NaN

# Finalmente, convierte a tipo 'category'
for column in cat_feats:
    cuestionario[column] = cuestionario[column].astype('category')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier

# Definir las columnas con valores nulos
cols_con_nulos = ['Tabaquismo', 'Drogas', 'Salud bucal', 'Actividad', 'Diabetes']

# 1. Imputación con Árbol de Decisión
cuestionario_dt = cuestionario.copy()
for col in cols_con_nulos:
    # Filtrar las filas donde el valor está nulo
    indices_nulos = cuestionario[cuestionario[col].isnull()].index
    # Crear un conjunto de entrenamiento excluyendo las filas con valores nulos en la columna
    data_train_dt = cuestionario.dropna(subset=[col])

    # Dividir en X (predictores) y y (variable objetivo)
    y_train = data_train_dt[col]
    X_train = data_train_dt.drop(cols_con_nulos, axis=1)

    # Predecir los valores nulos
    X_test = cuestionario.loc[indices_nulos].drop(cols_con_nulos, axis=1)
    dt_imputer = DecisionTreeClassifier()
    dt_imputer.fit(X_train, y_train)

    # Obtener las predicciones
    data_test_dt = dt_imputer.predict(X_test)

    # Asignar las predicciones a las filas con valores nulos en la columna original
    cuestionario_dt.loc[indices_nulos, col] = data_test_dt
```
